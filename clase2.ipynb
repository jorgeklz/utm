Método constructor

```python
### Programa para declarar metodos ###
class Perro :
  # Atributos
  especie = "mamifero"

  # Metodo constructor
  def __init__(self, nombre, genero):
    self.nombre = nombre
    self.genero = genero

  # Metodo para mostrar info de objetos
  def mostrar_info(self):
    print ("{} es {}".format (self.nombre , self.genero))

# Instanciacion del objeto perro
coqui = Perro("coqui","hembra")
rayo = Perro ("rayo","macho")

# Llamada a metodo para mostrar atributos del objeto
coqui . mostrar_info()
rayo. mostrar_info()
```

Agregar y modificar atributos de objetos:

```python
class Perro :
  # Metodo constructor
  def __init__ (self, nombre, crias):
    self. nombre = nombre
    self.crias = crias
  
  # Metodo para agregar propietario
  def agregar_propietario(self, propietario):
    self. propietario = propietario
    print (" Propietario de {} es {}".format(self.nombre, self. propietario))

  # Metodo para aumentar crias
  def aumentar_crias(self):
    self.crias+=1
    print ("Crias hoy", self. crias )


# Instanciacion del objeto perro
coqui = Perro("coqui ", 2)

# Agrego atributo propietario
coqui.agregar_propietario("Marian")
# Aumenta crias
coqui.aumentar_crias()
```

Realizar operaciones:

```python
class Perro:
  # Metodo constructor
  def __init__(self, nombre, crias):
    self. nombre = nombre
    self.crias = crias

  # Metodo para calcular gastos por crias
  def calcular_gasto(self):
    gasto_mensual = 20
    if self.crias >= 3:
      a_pagar = gasto_mensual * self. crias
    else:
      a_pagar = gasto_mensual * 1.5
    return a_pagar


# Instanciacion del objeto perro
coqui = Perro("coqui", 3)

# Calcular y mostrar gasto asociado
print("{} produce gastos por: {}".format(coqui.nombre, coqui.calcular_gasto()))
```
