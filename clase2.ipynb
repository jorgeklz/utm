## **Polimorfismo**

Crearemos dos clases distintas para usar con dos objetos distintos (Tiburon y Pez Payaso) pero con mismos métodos.

```python
class Tiburon():
    def nadar(self):
        print("El tiburon esta nadando.")
    def nadar_hacia_atras(self):
        print("El tiburon no nada hacia atras, pero puede hundirse.")
    def esqueleto(self):
        print("El esqueleto de tiburon esta hecho de cartilago.")

class PezPayaso():
    def nadar(self):
        print("El pez payaso esta nadando.")
    def nadar_hacia_atras(self):
        print("El pez payaso nada hacia atras.")
    def esqueleto(self):
        print("El esqueleto de pez payaso esta hecho de cartilago.")
```

Creamos los objetos

```python
tib1 = Tiburon()

pez1 = PezPayaso()
```

Crearemos **polimorfismo con métodos** con un bucle for que itera a través de una tupla de objetos. 

```python
for animal in (tib1, pez1):
    animal.nadar()
    animal.nadar_hacia_atras()
    animal.esqueleto()
```

Crearemos **polimorfismo con funciones**
